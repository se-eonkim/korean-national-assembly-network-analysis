```python
import pandas as pd
import networkx as nx
from community import best_partition
from collections import defaultdict

# 1. 엣지 데이터 읽기
edges_df = pd.read_csv('의원데이터/전체데이터/제20대_법안별_전체데이터__2018-02-13_이전_edge_무방향.csv')

# 2. 노드 데이터 읽기
nodes_df = pd.read_csv('의원데이터/전체데이터/제20대_법안별_전체데이터_node.csv')

# 3. 무방향 그래프 생성
G = nx.Graph()

# 3.1 노드 추가
for index, row in nodes_df.iterrows():
    G.add_node(row['Id'], Party=row['Party'])

# 3.2 엣지 추가 (중복 엣지 가중치 합산)
for index, row in edges_df.iterrows():
    if G.has_edge(row['Source'], row['Target']):
        G[row['Source']][row['Target']]['weight'] += 1
    else:
        G.add_edge(row['Source'], row['Target'], weight=1)

# 4. 정당 모듈러리티 계산
def calculate_party_modularity(graph, party_partition):
    """
    Calculate modularity for predefined party-based communities.
    """
    community_dict = defaultdict(list)
    for node, party in party_partition.items():
        community_dict[party].append(node)
    communities = list(community_dict.values())

    return nx.algorithms.community.quality.modularity(graph, communities, weight='weight')

# 정당별 커뮤니티 정의
party_partition = {node: data['Party'] for node, data in G.nodes(data=True)}

# 정당 모듈러리티 계산
party_modularity = calculate_party_modularity(G, party_partition)
print(f"정당 모듈러리티: {party_modularity:.4f}")

# 5. 최대 모듈러리티 계산
def calculate_max_modularity(graph):
    """
    Calculate maximum modularity using Louvain algorithm.
    """
    partition = best_partition(graph, weight='weight')  # Louvain 알고리즘으로 최적 커뮤니티 탐색
    community_dict = defaultdict(list)
    for node, community in partition.items():
        community_dict[community].append(node)
    communities = list(community_dict.values())

    return nx.algorithms.community.quality.modularity(graph, communities, weight='weight')

# 최대 모듈러리티 계산
max_modularity = calculate_max_modularity(G)
print(f"최대 모듈러리티: {max_modularity:.4f}")

# 6. 비율 계산
ratio = party_modularity / max_modularity if max_modularity != 0 else 0
print(f"비율: {ratio:.4f}")
```

```python
import pandas as pd
import networkx as nx
from collections import Counter
import numpy as np

# 1. 엣지와 노드 CSV 파일 목록 정의
edge_files = [
    '의원데이터/전체데이터/제20대_법안별_전체데이터__2018-02-13_이전_edge_무방향.csv',
    '의원데이터/전체데이터/제20대_법안별_전체데이터__2018-02-13_이후_edge_무방향.csv',
    '의원데이터/전체데이터/제20대_법안별_전체데이터_edge_무방향.csv',
    '의원데이터/전체데이터/제21대_법안별_전체데이터_edge_무방향.csv'
]

node_files = [
    '의원데이터/전체데이터/제20대_법안별_전체데이터_node.csv',
    '의원데이터/전체데이터/제20대_법안별_전체데이터__2018-02-13_이후_node.csv',
    '의원데이터/전체데이터/제20대_법안별_전체데이터_node.csv',
    '의원데이터/전체데이터/제21대_법안별_전체데이터_node.csv'
]

# 2. 협력 지수를 계산하는 함수 정의
def calculate_collaboration_index_undirected(graph, top_parties):
    """
    상위 5개 정당 간의 협력 지수를 계산합니다 (무방향 그래프).
    같은 정당 내의 협력은 제외합니다.
    """
    collaboration_indices = {}
    for i, party_a in enumerate(top_parties):
        for j, party_b in enumerate(top_parties):
            if i >= j:
                continue  # 같은 정당 및 중복 계산 제외
            # 각 정당에 속한 노드 가져오기
            nodes_a = [node for node, data in graph.nodes(data=True) if data['Party'] == party_a]
            nodes_b = [node for node, data in graph.nodes(data=True) if data['Party'] == party_b]

            sij_values = []
            for node_i in nodes_a:
                for node_j in nodes_b:
                    # c_ij 계산
                    if graph.has_edge(node_i, node_j):
                        c_ij = graph[node_i][node_j]['weight']
                    else:
                        c_ij = 0
                    # 차수 계산
                    d_i = sum([data['weight'] for _, _, data in graph.edges(node_i, data=True)])
                    d_j = sum([data['weight'] for _, _, data in graph.edges(node_j, data=True)])
                    sij = c_ij / (np.sqrt(d_i * d_j)) if d_i > 0 and d_j > 0 else 0
                    sij_values.append(sij)

            # 총 쌍의 수 계산
            total_pairs = len(nodes_a) * len(nodes_b)
            # P_AB 계산
            p_ab = sum(sij_values) / total_pairs if total_pairs > 0 else 0
            collaboration_indices[(party_a, party_b)] = p_ab

    return collaboration_indices

# 3. 각 네트워크에 대해 처리
for edge_file, node_file in zip(edge_files, node_files):
    # 3.1 엣지 데이터 읽기
    edges_df = pd.read_csv(edge_file)

    # 3.2 노드 데이터 읽기
    nodes_df = pd.read_csv(node_file)

    # 3.3 무향성 그래프 생성
    G = nx.Graph()

    # 3.4 노드 추가
    for index, row in nodes_df.iterrows():
        G.add_node(row['Id'],
                   Label=row['Label'],
                   Party=row['Party'],
                   Reselection=row['Reselection'],
                   District_Type=row['District_Type'],
                   Gender=row['Gender'],
                   Role=row['Role'])

    # 3.5 엣지 추가 (중복 엣지 가중치 합산)
    for index, row in edges_df.iterrows():
        if G.has_edge(row['Source'], row['Target']):
            # 이미 존재하는 엣지의 가중치에 추가
            G[row['Source']][row['Target']]['weight'] += 1
        else:
            # 새 엣지를 추가
            G.add_edge(row['Source'], row['Target'], weight=1)

    # 4. 정당별 커뮤니티 정의
    party_partition = {node: data['Party'] for node, data in G.nodes(data=True)}

    # 5. 정당별 노드 수 상위 5개 추출
    party_counts = Counter([data['Party'] for _, data in G.nodes(data=True)])
    top_5_parties = [party for party, _ in party_counts.most_common(5)]

    # 6. 협력 지수 계산
    collaboration_indices_undirected = calculate_collaboration_index_undirected(G, top_5_parties)

    # 7. 결과 출력
    print(f"엣지 파일 '{edge_file}'의 정당 간 협력 지수 (상위 5개 정당):")
    for (party_a, party_b), value in collaboration_indices_undirected.items():
        print(f"{party_a} - {party_b}: {value:.10f}")
    print("\n")
```
